```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
data_frame=pd.read_csv('Iris.csv')
```

```python
data_frame.head()
```

```python
data_frame.rename(columns={'SepalLengthCm': 'sep_len','SepalWidthCm': 'sep_wid', 'PetalLengthCm' : 'pet_len','PetalWidthCm': 'pet_wid'}, inplace=True)
```

```python
data_frame.describe()
```

```python
sns.pairplot(data_frame, hue='Species')
```

```python
dat=data_frame.values
x=dat[:,0:4]
y=dat[:,5]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=20)
```

```python
print(y_train)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr=LogisticRegression()
lr.fit(x_train , y_train)
```

```python
from sklearn.metrics import accuracy_score
```

```python
pred= lr.predict(x_test)
print(accuracy_score(y_test,pred))
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test,pred))
```

```python
new= np.array([[3,2,1,0.2]])
prediction=lr.predict(new)
print("Prediction of species is :"+ prediction)
```
